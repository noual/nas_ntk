## NAS-Bench-201
Let's all begin with our favorite search space, NAS-Bench-201.

```python
from naslib.search_spaces import nasbench201 as nb201
```

Plot all architectures

```python
from nas_201_api import NASBench201API
```

```python
dataset_api = get_dataset_api("nasbench201", 'cifar10')
real_nb201_api = NASBench201API("../naslib/data/NAS-Bench-201-v1_0-e61699.pth", verbose=False)
```

```python
from xautodl.models import get_cell_based_tiny_net
from naslib.search_spaces.nasbench201.conversions import convert_str_to_op_indices, convert_op_indices_to_naslib
params = []
accuracies = []

for i, arch_str in tqdm(enumerate(real_nb201_api)):
    indices = convert_str_to_op_indices(arch_str)
    objec = nb201.graph.NasBench201SearchSpace()
    convert_op_indices_to_naslib(indices, objec)
    naslib_results = objec.query(dataset='cifar10', metric=Metric.VAL_ACCURACY, dataset_api=dataset_api)
    id = real_nb201_api.query_index_by_arch(arch_str)
    info = real_nb201_api.query_meta_info_by_index(id, hp="200")
    nb_results = info.get_metrics("cifar10-valid", "x-valid")["accuracy"]
    config = real_nb201_api.get_net_config(id, 'cifar10-valid')  # obtain the network configuration for the n-th architecture on the CIFAR-10 dataset
    network = get_cell_based_tiny_net(config)  # create the network from configuration
    n_params = sum([np.prod(p.size()) for p in network.parameters()])
    params.append(n_params)
    accuracies.append(nb_results)
```

```python
sns.scatterplot(x=params, y=accuracies)
plt.xlabel("Number of parameters")
plt.ylabel("Validation accuracy")
plt.title("NAS-Bench-201")
plt.savefig("nas_bench_distribution.png")
plt.show()
```

```python
import random
import numpy as np
from nas_201_api import NASBench201API
from naslib.search_spaces.nasbench201.conversions import convert_str_to_op_indices, convert_op_indices_to_naslib, convert_naslib_to_str
from naslib.search_spaces.nasbench201.graph import NasBench201SearchSpace
from naslib.utils import get_dataset_api
from naslib.search_spaces.core.query_metrics import Metric

class RegularizedEvolution:
    def __init__(self, population_size, sample_size, dataset_api, real_nb201_api):
        self.population_size = population_size
        self.sample_size = sample_size
        self.dataset_api = dataset_api
        self.real_nb201_api = real_nb201_api
        self.population = []

    def initialize_population(self):
        for _ in range(self.population_size):
            arch_str = random.choice(self.real_nb201_api)
            self.population.append(arch_str)

    def mutate(self, parent_arch_str):
        parent_indices = convert_str_to_op_indices(parent_arch_str)
        child_indices = parent_indices.copy()
        edge = random.randint(0, len(parent_indices) - 1)
        available_ops = [i for i in range(len(NasBench201SearchSpace.OP_NAMES)) if i != parent_indices[edge]]
        child_indices[edge] = random.choice(available_ops)
        return convert_op_indices_to_naslib(child_indices, NasBench201SearchSpace())

    def evaluate(self, arch):
        arch_str = convert_naslib_to_str(arch)
        return arch.query(dataset='cifar10', metric=Metric.VAL_ACCURACY, dataset_api=self.dataset_api)

    def evolve(self, generations):
        self.initialize_population()
        for _ in range(generations):
            sample = random.sample(self.population, self.sample_size)
            parent_arch_str = max(sample, key=lambda arch_str: self.evaluate(convert_op_indices_to_naslib(convert_str_to_op_indices(arch_str), NasBench201SearchSpace())))
            child_arch = self.mutate(parent_arch_str)
            self.population.append(convert_naslib_to_str(child_arch))
            self.population.pop(0)

        best_arch_str = max(self.population, key=lambda arch_str: self.evaluate(convert_op_indices_to_naslib(convert_str_to_op_indices(arch_str), NasBench201SearchSpace())))
        return best_arch_str

# Example usage
evolution = RegularizedEvolution(population_size=50, sample_size=10, dataset_api=dataset_api, real_nb201_api=real_nb201_api)
best_architecture = evolution.evolve(generations=100)
print("Best architecture:", best_architecture)
```
